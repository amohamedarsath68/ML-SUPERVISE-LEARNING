```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler,PolynomialFeatures
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
#regression model
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
```

```python
#load datadet
df=pd.read_csv("data.csv")
print("shape:",df.shape)
print("columns:",df.columns.tolist())
print("\nmissing value:\n",df.isnull().sum())
print("\nduplicate Rows:",df.duplicated().sum())
```

```python
#drop irrelevent data
df=df.drop(['date','street','city','statezip','country'],axis=1)
```

```python
for col in df.select_dtypes(include=['int64','float64']).columns:
    df[col].fillna(df[col].median(),inplace=True)
for col in df.select_dtypes(include=['object']).columns:
    df[col].fillna(df[col].mode(),inplace=True)
```

```python
label_encoder={}
for col in df.select_dtypes(include=['object']).columns:
    le=LabelEncoder()
    df[col]=le.fit_transform(df[col])
    label_encoder[col]=le
```

```python
target_column="price"
X=df.drop(target_column, axis=1)
y=df[target_column]
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=42)
```

```python
def evaluate_model(y_true,y_pred):
    mae=mean_absolute_error(y_true,y_pred)
    mse=mean_squared_error(y_true,y_pred)
    rmse=np.sqrt(mse)
    r2=r2_score(y_true,y_pred)
    print("MAE:",mae)
    print("MSE:",mse)
    print("RMSE:",rmse)
    print("R2 SCORE:",r2)
```

```python
print("\n=====+ Linear Regression ====")
lr=LinearRegression()
lr.fit(x_train, y_train)
y_pred =lr.predict(x_test)
evaluate_model(y_test,y_pred)
```

```python
print("\n ==== Multiple Linear Regression ====")
mlr=LinearRegression()
mlr.fit(x_train,y_train)
y_pred=mlr.predict(x_test)
evaluate_model(y_test,y_pred)
```

```python
print("\n====polynomial Regression====")
poly=PolynomialFeatures(degree=2)
x_train_poly=poly.fit_transform(x_train)
x_test_poly=poly.transform(x_test)

poly_model=LinearRegression()
poly_model.fit(x_train_poly,y_train)
y_pred=poly_model.predict(x_test_poly)
evaluate_model(y_test,y_pred)
```

```python
print("\====Ridge Regression ====")
ridge=Ridge(alpha=1.0)
ridge.fit(x_train,y_train)
y_pred=ridge.predict(x_test)
evaluate_model(y_test,y_pred)
```

```python
#LASSO REGRESSION
print("\n==== Lasso Regression ====")
lasso= Lasso(alpha=0.01)
lasso.fit(x_train,y_train)
y_pred=lasso.predict(x_test)
evaluate_model(y_test,y_pred)
```

```python
print("\n=== Decision Tree Regression===")
dt=DecisionTreeRegressor(max_depth=5)
dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)
evaluate_model(y_test,y_pred)
```

```python
print("\n===Elasticnet Regression===")
elastic=ElasticNet(alpha=0.01,l1_ratio=0.05)
elastic.fit(x_train,y_train)
y_pred=elastic.predict(x_test)
evaluate_model(y_test,y_pred)
```

```python
print("\n===Random Forest Regressiom===")
rf=RandomForestRegressor(n_estimators=100,random_state=42)
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
evaluate_model(y_test,y_pred)
```

```python
print("\n===Support Vectors Regression===")
svr=SVR(kernel='rbf')
svr.fit(x_train,y_train)
y_pred=svr.predict(x_test)
evaluate_model(y_test,y_pred)
```

```python
print("\n=== K Nearest Neightbors Regression===")
knn=KNeighborsRegressor(n_neighbors=5)
knn.fit(x_train,y_train)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
evaluate_model(y_test,y_pred)
```
