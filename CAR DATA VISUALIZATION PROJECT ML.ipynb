```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("CAR DETAILS FROM CAR DEKHO.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#count call
plt.figure(figsize=(8,6))
sns.countplot(x='seller_type',data=df,palette='viridis')
plt.title('selling_type count plot')
plt.xlabel('seller_type')
plt.ylabel('count')
plt.show()
```

```python
plt.Figure(figsize=(8,6))
sns.histplot(df['year'].dropna(),bins=30,kde=True,color='green')
plt.title('Hisplot For Year')
plt.xlabel('year')
plt.ylabel('count')
plt.show()
```

```python
cars=df['owner'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(cars,labels=['first owner','second owner','third owner','fourth & above owner','test drie car'],
        autopct='%1.1f%%',colors=['black','lightblue','lightgreen','lightyellow','yellow'],
        startangle=180,explode=(0.05,0.05,0.05,0.05,0.05))
plt.title('pie chart for car ')
plt.show()
```

```python
cars=df['owner'].value_counts()
cars
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='seller_type',y='selling_price',data=df,palette='viridis')
plt.title('Boxplot for selling price')
plt.xlabel('Seller Type') 
plt.ylabel('Selling Price')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='year',y='km_driven',hue='owner',style='seller_type',data=df,palette='rainbow')
plt.title('year and km_driven with owner and seller_type')
plt.xlabel('year')
plt.ylabel
plt.legend(title='about car')
plt.show()
```

```python
df.info()
```

```python
num=df[['year','selling_price','km_driven']].corr()
num
```

```python
plt.Figure(figsize=(10,8))
sns.heatmap(num,annot=True,cmap='rainbow',center=0)
plt.title('correlation Hearmap of car dataset')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.violinplot(x='fuel',y='year',hue='seller_type',data=df,split=True,palette='rainbow')
plt.title('seller_type distribution with fuel and year')
plt.xlabel('fuel')
plt.ylabel('year')
plt.legend(title='survived')
plt.show()
```

```python
sns.kdeplot(data=df['year'].dropna(),color='skyblue',shade=True)
plt.title('year Distribution(kernal density estimation)')
plt.xlabel('year')
plt.ylabel('density')
plt.show()
```

```python
g=sns.FacetGrid(df,col='seller_type',row='owner',margin_titles=True)
g.map(sns.histplot,'year',bins=20,color='green')
g.fig.suptitle('age distribution by class and survial',y=1.03)
plt.show()
```

```python
sns.pairplot(df[['year','selling_price','km_driven','transmission']],
                 hue='transmission',palette='husl',
                 plot_kws={'alpha':0.6},diag_kind='kde')
plt.suptitle('pair plot of titanc variables',y=1.02)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.swarmplot(x='fuel',y='km_driven', hue='transmission',
              data=df,palette='rainbow',dodge=True)
plt.title('transmission Distribution bY fuel and Km_driven(swarm plot)')
plt.xlabel('fuel')
plt.ylabel('km_driven')
plt.legend(title='transmission')
plt.show()
```
